```python
# 1.Reverse a String
str1="python"
rev=''
for i in range(len(str1)-1,-1,-1):
    rev=rev+str1[i]
print(str1)
print(rev)
```

```python
# 2.	Check if a String is a Palindrome
str1='aabbaac'
rev=''
for i in range(len(str1)-1,-1,-1):
    rev=rev+str1[i]
if rev==str1:
    print('palindrome')
else:
    print('not palindrome')
```

```python
# 3.	Count the Number of Vowels in a String
str1='python'
vowels='aeiou'
count=0
for i in str1:
    if i in vowels:
        count+=1
print(count)
```

```python
# 7.	Check if a String Contains Only Digits
str1='python123'
digits='0123456789'
alphabets='abcdefghijklmnopqrstuvwxyz'
count=0
for i in str1:
    if i in digits and i not in alphabets:
        count+=1
    else:
        count=0
if count>0:
    print('yes')
else:
    print('no')
```
