```python
# 1.	Print Numbers From 1 to 10 Using a While Loop
i=1
while i<11:
    print(i,end=' ')
    i+=1
```

```python
# 2.	Calculate the Sum of Digits of a Given Number
num=1234
rem=0
sum=0
while num>0:
    rem=num%10
    sum=sum+rem
    num=num//10
print(sum)
```

```python
# 3.	Reverse a Number Using a While Loop
num=1234
rem=0
rev=0
while num>0:
    rem=num%10
    rev=rev*10+rem
    num=num//10
print(rev)
```

```python
# 4.	Print a Countdown From 10 to 1
i=10
while i>=1:
    print(i,end=' ')
    i-=1
```

```python
# 7.	Check if a Number is a Palindrome
num=121
rev=0
rem=0
temp=num
while num>0:
    rem=num%10
    rev=rev*10+rem
    num=num//10
if rev==temp:
    print('palindrome')
else:
    print('not palindrome')
```

```python
# 12.	Implement a Simple Menu-Driven Program Using While Loop
choice=int(input('enter choice: '))
a=5
b=5
while choice<=4:
    if choice==1:
        print(a+b)
        break
    elif choice==2:
        print(a-b)
        break
    elif choice==3:
        print(a*b)
        break
    else:
        print(a//b)
        break
```

```python
# 13.	Simulate a Basic Password Check That Limits to 3 Attempts
password=123
choice=int(input('enter password: '))
attempt=0
while attempt<4:
    attempt+=1
    if choice==password:
        print('success')
        break
    else:
        print('failure')
        break
```
