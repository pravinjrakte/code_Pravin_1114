```python
# 1.	Access Elements in a Tuple
tuple1=(1,2,3,4,5)
for i in tuple1:
    print(i,end=' ')
```

```python
# 2.	Check if an Element Exists in a Tuple
tuple1=(1,2,3,4,5)
element=5
count=0
for i in tuple1:
    if i==element:
        count+=1
if count>0:
    print('available')
else:
    print('not available')
```

```python
# 4.	Find the Index of an Element in a Tuple
tuple1=(1,2,3,4,5)
element=2
n=len(tuple1)
index=0
for i in range(0,n):
    if tuple1[i]==element:
        index=i
print(index)
```

```python
# 5.	Convert a List to a Tuple and Vice Versa
list1=[1,2,3,4,5]
tuple1=tuple(list1)
print(tuple1)
tuple2=(1,2,3,4,5,5,4,3)
list2=list(tuple2)
print(list2)
```

```python
# 7.	Concatenate Two Tuples
tuple1=(1,2,3,4,5)
tuple2=(1,2,3,5,6,7,8)
tuple3=tuple1+tuple2
tuple3
```

```python
# 8.	Check if Two Tuples are Identical
tuple1=(1,2,3,4,5)
tuple2=(1,2,3,5,6,7,8)
if tuple1==tuple2:
    print('equal')
else:
    print('not equal')
```

```python
# 9.	Sort a Tuple
tuple1=(1,2,3,4,53,4,52,65)
tuple1
n=len(tuple1)
for i in range(0,n):
    for j in range(i+1,n):
        if tuple1[i]>tuple1[j]:
            tuple1[j]=tuple1[i]
tuple1

```

```python
# 10.	Find the Length of a Tuple
tuple1=(1,2,3,4,53,4,52,65,4)
print(tuple1)
n=len(tuple1)
print(n)
```

```python
# 12.	Reverse a Tuple
tuple1=(1,2,3,4,53,4,52,65,4)
n=len(tuple1)
for i in range(n-1,-1,-1):
    print(tuple1[i],end=' ')
```

```python
# 14.	Iterate Over a Tuple and Print Each Element
tuple1=(1,2,3,4,53,4,52,65,4)
for i in tuple1:
    print(i,end=' ')
```
