```python
# 1.	Print All Elements in a List Using a For Loop
list1=[1,2,3,4,5]
for i in list1:
    print(i,end=' ')
```

```python
# 2.	Calculate the Sum of All Numbers in a List
list1=[1,2,3,4,5]
sum=0
for i in list1:
    sum=sum+i
print(sum)
```

```python
# 3.	Find the Factorial of a Given Number
num=5
fact=1
for i in range(1,num+1):
    fact=fact*i
fact
```

```python
# 4.	Print the Multiplication Table of a Given Number
num=2
for i in range(1,11):
    print(f"{num} * {i} = {num*i} ")
```

```python
# 5.	Print the Fibonacci Sequence Up to n Terms
n=5
a=0
b=1
print(a,end=' ')
print(b,end=' ')
c=0
for i in range(1,6):
    c=a+b
    print(c,end=' ')
    a=b
    b=c
```

```python
# # 6.	Count the Number of Even and Odd Numbers in a List
list1=[1,2,3,4,5]
even_count=0
odd_count=0
for i in list1:
    if i%2==0:
        even_count+=1
    else:
        odd_count+=1
print(even_count)
print(odd_count)
```

```python
# 7.	Reverse a String Using a For Loop
str1='python'
rev=''
n=len(str1)
for i in range(n-1,-1,-1):
    rev=rev+str1[i]
print(rev)
```

```python
# 8.	Find the Maximum and Minimum Values in a List
list1=[1,2,3,4,5]
min_element=list1[0]
max_element=list1[0]
for i in list1:
    if i<min_element:
        min_element=i
print(min_element)
for i in list1:
    if i>max_element:
        max_element=i
print(max_element)
```

```python
# 9.	Print All Prime Numbers Within a Given Range
first=1
last=50
count=0
for i in range(first,last+1):
    for j in range(2,last-1):
        if i%j==0:
            count+=1

```

```python
# 10.	Iterate Over a Dictionary and Print Each Key-Value Pair
student = {
    "name": "Pravin",
    "age": 21,
    "course": "Python"
}
for k,v in student.items():
    print(k,':',v)
```
