```python
# 1.	Access the Value Associated with a Key
student = {
    "name": "Pravin",
    "age": 22,
    "course": "Python",
    "marks": 85
}
print(student["name"])
```

```python
# 2.	Check if a Key Exists in a Dictionary
student = {
    "name": "Pravin",
    "age": 22,
    "course": "Python",
    "marks": 85
}
list1=[student.keys()]
list1
key='name'
for i in list1:
    if i == key:
        print('yes')
    else:
        print('no')

```

```python
# 3.	Iterate Over a Dictionary and Print All Key-Value Pairs
student = {
    "name": "Pravin",
    "age": 22,
    "course": "Python",
    "marks": 85
}
for k,v in student.items():
    print(student.keys(),student.values())
```
