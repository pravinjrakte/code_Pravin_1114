```python
# 1.	Check if a Number is Positive, Negative, or Zero
num=15
if num>0:
    print('positive')
elif num<0:
    print('negative')
else:
    print('zero')
```

```python
# 2.	Determine if a Person is Eligible to Vote Based on Age
age=18
if age>=18:
    print('eligible')
else:
    print('not eligible')
```

```python
# 3.	Check if a Year is a Leap Year
year=2003
if year%4==0 or year%400==0:
    print('leap')
else:
    print('normal')
```

```python
# 4.	Find the Largest of Three Numbers
a=10
b=20
c=15
if a>b and a>c:
    print('a largest')
elif b>a and b>c:
    print('b largest')
else:
    print('c largest')
```

```python
# 5.	Check if a Character is a Vowel or Consonant
char='b'
vowels='aeiou'
for i in char:
    if i in vowels:
        print('vowels')
    else:
        print('consonants')
```

```python
# 9.	Check if Two Numbers are Equal, Greater, or Lesser
num1=5
num2=10
if num1==num2:
    print('same')
elif num1>num2:
    print('greater')
else:
    print('smaller')
```

```python
# 10.	Check if a Number is Divisible by Both 3 and 5
num=15
if num%3==0 and num%5==0:
    print('divisible')
else:
    print('not divisible')
```

```python
# 11.	Implement a Simple Calculator Using if-elif-else Statements
num1=10
num2=5
char='+'
if char=='+':
    print(num1+num2)
elif char=='-':
    print(num1-num2)
elif char=='*':
    print(num1*num2)
else:
    print(num1//num2)
```
